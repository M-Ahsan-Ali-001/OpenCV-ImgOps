```python
# imporitng libraries
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# read
mrThief = cv2.imread("lupin.jpg")
```

```python
# Size Setup - Conversion - Display
#not open cv read the image in BGR format we must have to convert to RGB
plt.figure(figsize= (10,10))
plt.imshow(cv2.cvtColor(mrThief, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Intializing the pointer the orignal image means that we are dublicating we are just making a pointer 
pointer = mrThief 
```

```python
#printing memory address of the image
print("Does pointer and mrThief have same memory address :",id(pointer)==id(mrThief) ,"\nMemory address:",id(pointer))
```

```python
# saving the image in a new memory address

dublicate = mrThief.copy()

```

```python
# Changing R and G channel of mrThief to zero
mrThief[:,:]=0
#mrThief[R,G,B]
```

```python
# Printing two photo after changing thier R ang G channel to zero
plt.figure(figsize=(10,10))
plt.subplot(131)#(rows , columns , index)
plt.imshow(cv2.cvtColor(mrThief, cv2.COLOR_BGR2RGB))
plt.title("mrThief")
plt.subplot(132)
plt.imshow(cv2.cvtColor(pointer,cv2.COLOR_BGR2RGB))
plt.title("pointer")

plt.subplot(133)
plt.imshow(cv2.cvtColor(dublicate,cv2.COLOR_BGR2RGB))
plt.title("dublicate")
plt.show
```

<h1 style="text-align:center">Fliping Images</h1>


```python
reader = cv2.imread("lupin.jpg")
```

```python
height,width,channels=reader.shape
print('height|width|channels:',height,width,channels)
```

```python
# making a 3 dimesnional array with type uint8= range is between(0-255)
img_fliped=np.zeros((height,width,channels),dtype=np.uint8)
```

```python
#storing rows of orignal in down to top order in img_fliped
for i,row in enumerate(reader):
    img_fliped[height-1-i,:,:]=row
    
```

```python
#printing rotated img
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img_fliped,cv2.COLOR_BGR2RGB))
plt.show()
```

<h1 style="text-align:center">Flip</h1>

```python
for flipit in [0,1,-1]:
    imgFlipped = cv2.flip(reader,flipit)
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(imgFlipped, cv2.COLOR_BGR2RGB))
    plt.show()
```

<h1 style="text-align:center">Cropping image</h1>

```python
# image croping
# [ top-down , left- right , : ]
upper=500
lower=650
reader = cv2.imread("lupin.jpg")
crop_it=reader[:,upper: lower,:]
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(crop_it,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# making a reactange on the image
#pt1(left,upper) ,pt2(right, lower)

img_rect=np.copy(reader)
cv2.rectangle(img_rect, pt1=(500,20) ,pt2 =(650,330),color=(0,0,255),thickness=3)
plt.imshow(cv2.cvtColor(img_rect,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#put TexT
#org=(left,top)
img_rect=np.copy(reader)
cv2.putText(img=img_rect,text="Lupin",org=(10,200) , color=(255,255,255),fontFace=4,fontScale=5,thickness=2)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img_rect,cv2.COLOR_BGR2RGB))
plt.show()
```
